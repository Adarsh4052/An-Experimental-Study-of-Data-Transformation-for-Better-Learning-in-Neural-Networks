```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure

from sklearn.externals import joblib
```

```python
#save the model  and results for further analysis
cnn_3 = joblib.load(r"results\acc_cnn_3_valid" )  
cnn_5 = joblib.load(r"results\acc_cnn_5_valid")

cnn_3_same = joblib.load(r"results\acc_cnn_3_same" )  
cnn_5_same = joblib.load(r"results\acc_cnn_3_same")

mlp = joblib.load(r"results\acc_mlp")
```

```python
figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')

plt.plot(cnn_3)
plt.plot(cnn_3_same)
plt.plot(cnn_5)
plt.plot(cnn_5_same)
plt.plot(mlp)

plt.title('Models Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['CNN 3','CNN 3 SAME', 'CNN 5' ,'CNN 5 SAME', 'MLP' ], loc='lower right')
plt.show()
```

```python
siamese_mlp_val_acc = joblib.load(r"results\siamese_mlp_val_acc" )  
siamese_cnn_val_acc = joblib.load(r"results\siamese_cnn_val_acc")
```

```python
figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')

plt.plot(siamese_cnn_val_acc)
plt.plot(siamese_mlp_val_acc)

plt.title('Models Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend([ 'Siamese CNN' ,'Siames MLP' ], loc='lower right')
plt.show()
```
